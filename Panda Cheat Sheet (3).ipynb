```python
# merge-adding a column to the file that has the average horespower for the number of cylinders
# calculate cylinder the mean of horsepower by cylinder
result=pd.DataFrame(df.groupby(['cylinders'])['horsepower'].aggregate('mean'))
result.head()
```

```python
# in the original file we want to add a column that has the average horespower for that row-vlookup on cylinders
new_df=pd.merge(df,result,how='left',left_on=['cylinders'],right_on=['cylinders'])
new_df.head()
```

```python
# replace null values in horsepower by replacing the null values with the average horsepower of cylinder

df.groupby(['cylinders'])['horsepower'].transform('mean').unique()
```

```python
# replace
df['horsepower']=df['horsepower'].fillna(df.groupby(['cylinders'])['horsepower'].transform('mean'))
```
